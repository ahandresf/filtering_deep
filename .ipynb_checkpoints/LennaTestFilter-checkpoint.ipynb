```python
#!/bin/usr/env python
import scipy
from scipy import ndimage, misc, signal
import skimage
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import matplotlib.cm as cm
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.ticker import LinearLocator, FormatStrFormatter
```

```python
#getting image
#image_file=('Lenna.png')
image_file=('lenna_unif.png')
img = mpimg.imread(image_file)
```

```python
print(type(img[0,0,0]))
```

```python
sig=img
#adding noise
sig_nse=skimage.util.random_noise(sig, mode='gaussian', seed=None, clip=True)
```

```python
fil_sig=signal.wiener(sig_nse, mysize=None, noise=None) #filter signal
```

```python
#PLOTTING
#Space domain s and s+noise
fig_t, axs_t = plt.subplots(1,3)
axs_t[0].imshow(sig,cmap=cm.gray)
axs_t[1].imshow(sig_nse,cmap=cm.gray)
axs_t[2].imshow(fil_sig,cmap=cm.gray)
```

```python
filt_original=signal.wiener(sig, mysize=None, noise=None) #filter signal
print(type(filt_original[0,0,0]))
```

```python
I=255*(filt_original-filt_original.min())/(filt_original.max()-filt_original.min())
I=I.round()
I = I.astype(np.uint8)
```

```python
#PLOTTING
#Space domain s and s+noise
fig_t, axs_t = plt.subplots(1,2)
axs_t[0].imshow(sig,cmap=cm.gray)
axs_t[1].imshow(I,cmap=cm.gray)
```

```python
#fourier transform
sig_f=scipy.fftpack.fft2(sig) #signal in frequency domain
sig_nse_f=scipy.fftpack.fft2(sig_nse) #signal+noise in frequency domain
fil_sig_f=scipy.fftpack.fft2(fil_sig)
```
